## Fish weight perdiction

 We have a dataset containing different fish species along with their physical characteristics, such as vertical length, diagonal length, height, and width. Our goal is to predict the weight of a fish based on these characteristics. To achieve this, we will use the Linear Regression method to analyze how the fishâ€™s weight is related to its physical features.

```python
import pandas as pd
```

```python
fish = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Fish.csv')
```

```python
fish.head()
```

```python
fish.info()
```

```python
fish.describe()
```

```python
fish.columns
```

```python
X = fish[['Category','Height', 'Width', 'Length1','Length2', 'Length3']]
y = fish['Weight']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=59)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
